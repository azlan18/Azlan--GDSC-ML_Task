**Importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Loading & preprocessing dataset**

```python
dataset = pd.read_csv('IRIS.csv')
dataset.head()
#Selecting the 'sepal_length' and 'sepal_width' features from the dataset
sepal_data = dataset[['sepal_length', 'sepal_width']]
print(sepal_data)
```

**Prior to clustering**

```python
plt.scatter(sepal_data['sepal_length'], sepal_data['sepal_width'])
```

**Apply K-Means Clustering**

```python
from sklearn.cluster import KMeans
# Create a KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, init = 'k-means++',random_state=0)
# Fit the model to the data (i.e., perform clustering)
kmeans.fit(sepal_data)
```

```python
# Get the cluster assignments for each data point
cluster_labels = kmeans.labels_
```

**Visualize the Clusters**

```python
# Plot the data points with cluster assignments
plt.scatter(sepal_data['sepal_length'], sepal_data['sepal_width'], c=cluster_labels)
plt.xlabel('Sepal Length ')
plt.ylabel('Sepal Width ')
plt.title('K-Means Clustering of Iris Dataset (Sepal Length vs. Sepal Width)')
plt.show()
```
